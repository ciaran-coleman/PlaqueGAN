```python
import numpy as np
from tqdm import tqdm

import pandas as pd

import os

from glob import glob

import shutil
import json
```

```python
def get_ucd_df(csv_path):
    df_all = pd.read_csv(csv_path)
    df_ucd = df_all[df_all['sourcetile'].str.contains('UCD')]
    df_ucd = df_ucd.rename(columns={"Unnamed: 0":"id"})
    df_ucd = df_ucd.drop(columns=['username','sourcetile'])
    return df_ucd
```

```python
df_ucd = get_ucd_df(csv_path)
```

```python
# function to copy over consensus UCDavis data to test folder
def copy_over_test_data(source_dir, target_dir, df_ucd):
    source_paths = [source_dir + '/' + img_path for img_path in df_ucd['imagename']]
    target_paths = [target_dir + '/' + img_path for img_path in df_ucd['imagename']]
    source_wsi = list(df_ucd['source'])
    for source_path, target_path, source in tqdm(zip(source_paths, target_paths, source_wsi), total=len(source_paths)):
        os.makedirs(os.path.join(target_dir,source),exist_ok=True)
        shutil.copy(source_path, target_path)
#     print(source_paths)
#     print(target_paths)
```

```python
source_dir = 'D:/ucl_masters_data/project/consensus_data/blobs'
target_dir = 'C:/Users/ciara/Documents/ucl_masters/modules/Project/Plaquebox/plaquebox-paper-master/data/tiles/hold-out'
copy_over_test_data(source_dir, target_dir, df_ucd)
```
